#### **Illustration of the problem and approach:**

**A:** Models that perform well on 1-step prediction (i.e., prediction order 0, in <font color='red'>red</font>), commonly fail in higher orders reconstruction (e.g., order-T , in <font color='lightgreen'>green</font>).

**B:** LINOCS (e.g., for training order $K = 3$) integrates weighted multi-step reconstructions for all $k = 0 . . . K$.
orders. It adapts the weights of these reconstruction orders to prioritize minimizing large errors at lower orders before addressing higher orders. The system gradually increases the weight of the most effective
lookahead reconstruction until convergence conditions are met.

**C:** LINOCS improves long-term reconstruction during training iterations, such that the predictions during training are expected to increasingly align
with the real system. <font color='blue'>Blue</font> color indicates the real system values. <font color='lightgreen'>Green</font> color indicates full-lookahead predictions across different iterations.

**D:** Weights of different lookahead training orders ($k$). $w_k : \mathbb{R}^2  \mathbb{R}$ is a function of the order ($k$) and the $k$-th order
reconstruction error ($e$).
*   *Top:* Illustration of an exemplary effect of $e$ on $w$ when fixing $k$.
*   *Bottom:* Illustration of an exemplary effect of $k$ on $w$ when fixing $e$.

# Example: Running Linear LINOCS with white noise addition

### 1) Create example linear data

````python
type_noise = 'w' # can be w (for white noise); s (for sinus noise); d (for 3-D case white noise).
addi = 'example_%s'%type_noise

np.random.seed(0)
"""
PARAMETERS
"""
dim = 2
noise_level = 0.3
T = 500
num_linocs_orders = 6

"""
create ofesst, transition matrices (F_lin), and observations (x, starting from x0)
"""
offset = np.random.rand(dim, 1)
F_lin =  create_rotation_mat(theta = 0.2, axes = 'x', dims = dim)
x0 = np.random.rand(dim, 1)
x = propagate_dyn_based_on_operator(x0, np.dstack([F_lin]*T), T,  offset = offset)

"""
add noise
"""
x_noisy = x + np.random.randn(*x.shape)*noise_level
x_noisy[:, 0] = x0.flatten()

x_minus_ones = np.vstack([x_noisy[:, :-1], np.ones((1, x.shape[1] - 1))])
x_no_offset = propagate_dyn_based_on_operator(x0, np.dstack([F_lin]*T), T)
````

````python
# @title Set Training and Prediction Orders to Test
orders_train =  [5, 10, 30,80]
orders_predict = [2, 10, 20, 80, 150, 200, T]
````

````python
# @title re-Train and Predict under Different Training and Prediction Orders

K_b_or = 20
func_w = np.exp
sigma_w = 0.01

recos = {}
recos_from_real =  {}
recos_one_step =  {}

for order_train in orders_train:
    #order_train = int(order_train)
    K_b = np.min([order_train, K_b_or])
    max_ord = order_train
    w = func_w(np.arange(max_ord)*sigma_w)#np.arange(max_ord)[::-1]*sigma_w #func_w(np.arange(max_ord)*sigma_w)#[::-1]#**4#[::-1]
    sigma_w_b = 1#0.5
    w_b = func_w(np.arange(K_b)*sigma_w_b)
    if K_b > 5:
        w_b[:5] = 0
    opt_A, opt_b,b_hats = train_linear_system_opt_A(x_noisy, max_ord, Bs_main = [], constraint = [], w_offset = True, cal_offset=True, weights=w, infer_b_way='each', K_b= K_b, w_b = w_b
                                  )
    offset_hat_opt = np.mean(opt_b, 1)
    rec1_hat_opt = propagate_dyn_based_on_operator(x0,opt_A, offset = offset_hat_opt, max_t = T)
    recos[order_train] = {}
    recos_from_real[order_train] = {}

    for order_predict in  orders_predict:
        #order_predict = int(order_predict)
        if order_predict == T:
            reco_now = rec1_hat_opt
            reco_now_from_real = rec1_hat_opt#propagate_dyn_based_on_operator(x0, opt_A, offset = offset_hat_opt, max_t = T)
        else:
            reco_now  = k_step_prediction_linear(x_noisy, opt_A, order_predict, store_mid = True, t = -1, offset = offset_hat_opt, verbose = False)
            reco_now_from_real  = k_step_prediction_linear(x, opt_A, order_predict, store_mid = True, t = -1, offset = offset_hat_opt, verbose = False)
        recos[order_train][order_predict] = reco_now
        recos_from_real[order_train][order_predict] = reco_now_from_real
        if order_predict not in recos_one_step:
            if order_predict == T:
                recos_one_step[order_predict] = propagate_dyn_based_on_operator(x0,F_1step, offset = b_1step, max_t = T)
            else:
                recos_one_step[order_predict] = k_step_prediction_linear(x, F_1step, order_predict, store_mid = True, t = -1, offset = b_1step, verbose = False)   #propagate_dyn_based_on_operator(x0,F_1step, offset = b_1step, max_t = T)
````

````python
# @title Plot 1-step vs. LINOCS for Multiple Training and Prediction Orders

fig, axs = plt.subplots(len(orders_train), len(orders_predict), figsize= (7*len(orders_predict), 7*len(orders_train)))
[[plot_2d(recos_from_real[orders_train[train_num]][orders_predict[pred_num]], ax = ax_i, params_plot={'lw':lw, 'alpha':0.6,'color':colors['LINOCS']}) for pred_num, ax_i in enumerate(ax)] for train_num, ax in enumerate(axs)]
[[plot_2d(recos_one_step[orders_predict[pred_num]], ax = ax_i, params_plot={'lw':lw, 'alpha':0.6,'color':colors['1 step']}) for pred_num, ax_i in enumerate(ax)] for train_num, ax in enumerate(axs)]
[plot_2d(x_noisy, params_plot={'color':'gray', 'ls':'-', 'alpha':0.8}, ax = ax_i) for ax_i in np.array(axs[0,0]).flatten()]


mid = int(len(orders_train)/2)
[axs[0,i].set_title('$K_{pred}=%d$'%(el), fontsize = title_size) for i,el in enumerate(orders_predict)]
[axs[i,0].set_ylabel('$K_{tra}=%d$'%(el) , fontsize = title_size) for i,el in enumerate(orders_train)]
[remove_edges(ax_i) for ax_i in axs.flatten()];
[ax_i.set_aspect('equal', adjustable='datalim') for ax_i in axs.flatten()]
````
